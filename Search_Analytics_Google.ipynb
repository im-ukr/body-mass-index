```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
Trending_topics = TrendReq(hl='en-US', tz=360)
```

```python
kw_list=["Generative AI"]
Trending_topics.build_payload(kw_list,cat=0, timeframe='today 12-m')
```

```python
Trending_topics.build_payload(kw_list=["Generative AI"],
							cat=0, timeframe='today 12-m')
data = Trending_topics.interest_over_time()
data = data.sort_values(by="Generative AI", ascending = False)
data = data.head(10)
print(data)
```

```python
from pytrends.request import TrendReq
import pandas as pd

pytrends = TrendReq()

kw_list = ["Generative AI"]

pytrends.build_payload(kw_list, cat=0, timeframe='2023-01-01 2023-02-01', geo='', gprop='')

data = pytrends.interest_over_time()

# Sort and get top 10 rows
if 'isPartial' in data.columns:
    data = data.drop(columns=['isPartial'])

data = data.sort_values(by="Generative AI", ascending=False)
data = data.head(10)

# Print data
print(data)
```

```python
data = Trending_topics.interest_by_region()
data = data.sort_values(by="Generative AI", ascending = False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName', y='Generative AI', figsize=(10,5), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```
